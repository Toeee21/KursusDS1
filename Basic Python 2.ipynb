Buatlah program dengan ketentuan sebagai berikut:
- Jika nilai tes di atas 60, siswa lulus ujian
- Jika nilai ujian antara 40-60, siswa diwajibkan untuk remedial
- Sedangkan jika nilai ujiannya di bawah 40, maka siswa tidak lulus ujian

```python
# 9. Buatlah program berdasarkan penjelasan diatas
score = float(input('Masukkan Nilai = '))
if (score > 60):
    print('Selamat Anda Lulus Ujian')
elif(score >= 40):
    print('Anda harus mengambil Remedial!')
else:
    print('Anda gagal dalam Tes!')
```

Anda sedang bekerja di NASA dan NASA membutuhkan program untuk menghitung mundur roketnya


```python
# 10. Buat program hitung mundur dari 10 hingga 0
countdown = 10
while (countdown >= 0):
    print(countdown)
    countdown = countdown - 1
print('Rocket Launched!!!')
```
